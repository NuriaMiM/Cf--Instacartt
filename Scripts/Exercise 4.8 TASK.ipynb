```python
# Importing libraries
import numpy as np
import pandas as pd
import os
```

```python
# Importing data
path = r'C:\Users\nmiqu\Desktop\DATA\CareerFoundry\Data Analytics Immersion\April-2024 Instacart Baket Analysis'
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_merge_derived.pkl'))
```

```python
ords_prods_merge.groupby('department_id').agg({'order_number' : ['mean']})
```

The result vary slightly from the subset we explored during the exercise, some departments are still very close (fro example department 1) but others do vary up to 3 points (for example department 21)

```python
# Step 4
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] >= 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] < 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'new customer'
```

```python
ords_prods_merge[['user_id', 'max_order', 'loyalty_flag']].head(20)
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean', 'min', 'max']})
```

These statistics let us know that although al type of clients buy products from all price ranges, new customers are the ones that, based on the mean, spend on higher priced items, followed by regular and then loyal customers

```python
# Step 6
ords_prods_merge['mean_spent'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_spent'] < 10, 'spender_type'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_spent'] >= 10, 'spender_type'] = 'High spender'
```

```python
ords_prods_merge['spender_type'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'mean_spent', 'spender_type']].head()
```

```python
# Step 7
ords_prods_merge['frequency_of_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.loc[ords_prods_merge['frequency_of_order'] > 20, 'frequency_type'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['frequency_of_order'] > 10) & (ords_prods_merge['frequency_of_order'] <= 20), 'frequency_type'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['frequency_of_order'] <= 10, 'frequency_type'] = 'Frequent customer'
```

```python
ords_prods_merge['frequency_type'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'frequency_of_order', 'frequency_type']].head()
```
