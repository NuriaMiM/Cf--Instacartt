```python
# Import Instacart data
ords_prods = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_merge_grouped.pkl'))
```

```python
ords_prods.dtypes
```

I identified that I won't be able to merge the dataframes using the user_id column because right now, in the ords_prods dataframe it is set as a 'float64', I would first have to transform it into an integer and then into a string. To be able to do the first transformation, I will have to drop null values, since Python doesn't allow me to go from float to integer if there are null values in a column. Taking a look at the previous list, I decided I will drop those 11 rows that have consisting missing values in different columns, this doesnt signify a big loss of data and will enable me to continue with the task.
I will also change all the floats that should be integers, this way I will prevent future errors like this one.

```python
ords_prods.dropna(subset = ['user_id'], inplace = True)
```

```python
ords_prods['user_id'] = ords_prods['user_id'].astype('int32')
ords_prods['order_id'] = ords_prods['order_id'].astype('int32')
ords_prods['order_number'] = ords_prods['order_number'].astype('int32')
ords_prods['orders_day_of_week'] = ords_prods['orders_day_of_week'].astype('int32')
ords_prods['days_since_prior_order'] = ords_prods['days_since_prior_order'].astype('int32')
ords_prods['reordered'] = ords_prods['reordered'].astype('int32')
```

```python
# Finally changing 'user_id' dtype to string to match the other set
ords_prods['user_id'] = ords_prods['user_id'].astype('str')
```

```python
# Dropping the '-merge' column to avoid confusion with the new one
ords_prods.drop('_merge', axis = 1, inplace = True)
```

```python
# Merging both datasets
cust_ords_prods = ords_prods.merge(customers, on = 'user_id', indicator = True)
```
